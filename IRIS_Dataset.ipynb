# Strip plot

```python
fig = plt.gcf()
fig.set_size_inches(10,7)
fig=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='green',size=8,palette='winter')
plt.show()
```

```python
# combined boxplot and stripplot
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxplot(x='Species',y='SepalLengthCm',data=iris)
fig=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='green',size=8,palette='winter')
plt.show()
```

```python
#...
# The jitter parameter in seaborn's stripplot function is used to add a small amount
#...

ax=sns.boxplot(x='Species',y='SepalLengthCm',data=iris)
ax=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='green',size=8,palette='winter')

boxtwo=ax.artists[2]
boxtwo.set_facecolor('yellow')
boxtwo.set_edgecolor('black')
boxthree=ax.artists[1]
boxthree.set_facecolor('red')
boxthree.set_edgecoloe('black')
boxthree=ax.artisist[0]
boxthree.set_facecolor('green')
boxthree.set_edgecolor('black')
plt.show()
```

# Heat Map

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Plot the heatmap
fig = plt.gcf()
fig.set_size_inches(10, 7)
sns.heatmap(iris_df.corr(), annot=True, cmap='cubehelix', linewidths=1, linecolor='k', square=True, mask=False, vmin=-1)
plt.show()
```
